```python
class Graph:
    def __init__(self, b, c, a, d):
        self.a, self.b, self.c, self.d = a, b, c, d
        self.GraphN = []
        self.df = []
    def createNode(self):
        self.obj = eval('Multi')(self.b,self.c)
        self.GraphN.append([self.b, self.obj,self.b])
        self.GraphN.append([self.c, self.obj,self.c])
        
        self.placeH = eval('PHolderOne')(self.obj.val)
        self.GraphN.append([self.obj, self.placeH,self.obj.val])
        
        self.obj = eval('Add')(self.a, self.placeH.val)
        self.GraphN.append([self.placeH,self.obj,self.placeH.val])
        self.GraphN.append([self.a, self.obj,self.a])
        
        self.placeH = eval('PHolderTwo')(self.obj.val)
        self.GraphN.append([self.obj, self.placeH,self.placeH.val])
        
        self.obj = eval('Multi')(self.placeH.val,self.d)
        self.GraphN.append([self.placeH, self.obj,self.placeH.val])
        self.GraphN.append([self.d, self.obj,self.d])
        self.GraphN.append([self.obj, eval('Result'),self.obj.val])
        print(self.GraphN)
        self.draw(self.GraphN)
        self.createFrame()
    def createNodes(self, dataF):
        for i in range(len(dataF)):
            if type(dataF[i][0]) is int:
                    self.drawManager.append([dataF[i][0],eval(dataF[i][-1]),dataF[i][1]])
            else :
                obj = eval(dataF[i][-1])(8)
                self.drawManager.append([eval(dataF[i][0]),obj,0])
    def createFrame(self):
        self.df = pd.DataFrame(self.GraphN, columns = ['Variable', 'Operation' , 'Weights'])
    def draw(self, elist , color = 'lightblue'):
        """
            Usage of networkx for visualisation
        """
        G = nx.DiGraph()
        G.add_weighted_edges_from(elist)
        plt.figure(figsize=(20,10))
       
        pos = nx.spring_layout(G)  # positions for all nodes
        nx.draw(G, pos, node_size=900, node_color=color, font_size=10, with_labels=True)
        nx.draw_networkx_edge_labels(G, pos, font_size=10,  edge_labels = nx.get_edge_attributes(G,'weight'))
```

```python
class Add:
    def __init__(self,X,Y):
        self.val = X + Y
        
class Multi:
    def __init__(self,X,Y):
        self.val = X * Y
        
class PHolderTwo:
    def __init__(self,X):
        self.val = X
class PHolderOne:
    def __init__(self,X):
         self.val = X
class Result:
    def __init__(self,X):
        pass
```
